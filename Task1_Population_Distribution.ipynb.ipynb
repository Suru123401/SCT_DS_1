```python
import matplotlib.pyplot as plt
#Data
categories = ['Male','Female']
values = [520,480]
#create bar chart
plt.figure(figsize=(6,4))
plt.bar(categories,values)

#Titles and labels
plt.title('Population Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

#Save chart
plt.savefig('population_distribution.png')

#Show chart
plt.show()
```
